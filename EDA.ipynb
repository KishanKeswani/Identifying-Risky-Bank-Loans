```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('credit.csv')
df[df['months_loan_duration']<5]
df[df['default']=='yes']
# df['default'].where(df['default']=='yes').dropna()
```

```python
y = df['default']
y=pd.get_dummies(y)
x=df.drop('default',1)
x=pd.get_dummies(x)
# y=y.map({'yes':1,'no':0})
x
```

```python
# splitting into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(x,y,random_state=42)
# using min_impurity_decrease (pruning) for feature engineering
classifier = DecisionTreeClassifier(min_impurity_decrease=0.005)
#training data
classifier.fit(X_train,Y_train)
```

```python
# predicting data
pred=classifier.predict(X_test)
d = pd.DataFrame(pred)
d=d[0]
print(Y_test)
d=d.map({1.0:'no',0.0:'yes'})
d #output
```

```python
# Visualizing decision tree
dotfile = open("D:/dtree2.dot", 'w')
tree.export_graphviz(classifier, out_file = dotfile, feature_names = X_test.columns)
dotfile.close()
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(classification_report(Y_test, pred))  
```
